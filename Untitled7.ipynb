# Exploratory data analysis for Chipotle data


## import data

```python
import pandas as pd
import numpy as np
import ssl
from collections import Counter
import matplotlib.pyplot as plt
ssl._create_default_https_context = ssl._create_unverified_context
#show all rows
pd.set_option('display.max_columns', None)
#show all columns
pd.set_option('display.max_rows', None)
pd.set_option('max_colwidth',100)

```

```python
url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv'

chipo = pd.read_csv(url, sep='\\t',engine='python')

#See the first 10 entries
print(chipo.head(10))
```

```python
chipo.item_price=chipo.item_price.apply(lambda x:float(x[1:-1]))
chipo.item_price.dtype  
chipo.describe()
```

## the price of each item?

```python
# need some techniques--first delete the duplicates then find the price where quantity is 1
chipo_cleaned=chipo.drop_duplicates(['item_name','quantity'])
chipo_one=chipo_cleaned[chipo_cleaned.quantity==1]
price_per_item = chipo_one[['item_name', 'item_price']]
price_per_item 
```

## Create a histogram of the top 5 items bought

```python
chipo_item=chipo.groupby('item_name').sum().sort_values(['quantity'],ascending=False)
chipo_item=chipo_item['quantity']
chipo_item=chipo_item[0:5]
# create the plot\n",
chipo_item.plot(kind='bar')



#bar  hist bar  pie

# Set the title and labels
plt.xlabel('Items')
plt.ylabel('Price')
plt.title('Most ordered Chipotle Items')
```

## Create a scatter plot with the number of items orderered per order price

```python
chipo['order_price']=chipo.quantity*chipo.item_price
orders = chipo.groupby('order_id').sum()
plt.scatter(x = orders.quantity, y = orders.order_price, s = 50, c = 'green')

# Set the title and labels

plt.xlabel('Items ordered')
plt.ylabel('Order Price')
plt.title('Number of items ordered per order price')
plt.ylim(0)
plt.show()
```
